```python
import folium
import json
import time
from IPython.display import display, clear_output
from folium.plugins import HeatMap
 
MAP_CENTER = [52.2297, 21.0122]
 
 
def load_metrics():
    with open("metrics.json", "r") as f:
        return json.load(f)
 
 
def show_dashboard():
 
    metrics = load_metrics()
 
    print("🚕 TAXI FLEET LIVE DASHBOARD\n")
    print(f"Available: {metrics['available_taxis']}")
    print(f"Occupied: {metrics['occupied_taxis']}")
    print(f"Revenue: {metrics['fleet_revenue']:.2f} zł")
    print(f"Rides: {metrics['completed_rides']}")

    print("\n🏆 TOP 10 TAXI\n")

    for i, taxi in enumerate(metrics["top_taxis"], start=1):
    
        print(
            f"{i}. Taxi {taxi['taxi_id']} | "
            f"Revenue: {taxi['earnings']:.2f} zł | "
            f"Distance: {taxi.get('distance_km', 0):.2f} km | "
            f"Rides: {taxi.get('completed_rides', 0)}"
        )
 
    m = folium.Map(
        location=MAP_CENTER,
        zoom_start=12
    )
 
    positions = metrics.get("positions", [])

    demand_points = metrics.get(
        "demand_points",
        []
    )
 
    for t in positions:
 
        color = "green" if t["status"] == "available" else "red"
 
        earnings = t.get("earnings", 0.0)
 
        folium.CircleMarker(
            location=[t["lat"], t["lon"]],
            radius=5,
            color=color,
            fill=True,
            fill_opacity=0.7,
            popup=f"""
            Taxi ID: {t['taxi_id']}<br>
            Status: {t['status']}<br>
            Revenue: {earnings:.2f} zł<br>
            Distance: {t.get('distance_km', 0):.2f} km<br>
            Rides: {t.get('completed_rides', 0)}
            """
        ).add_to(m)

    if len(demand_points) > 0:
        HeatMap(
            demand_points,
            radius=18,
            blur=12
        ).add_to(m)
    
    display(m)
 
 
# LIVE DASHBOARD
while True:
 
    clear_output(wait=True)
 
    try:
        show_dashboard()
 
    except Exception as e:
        print("Dashboard error:", e)
 
    time.sleep(3)
```
